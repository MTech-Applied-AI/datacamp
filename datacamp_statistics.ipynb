### Measures of Center (Summary Statistics)

#### it represents the center point.

Where is the **center of the data**?

* **Mean** - Average
* **Median** - 50% of data is lower and 50% of data is higher than this
* **Mode** - most frequent value in data (often used for **categorical** values)

- **Mean** is more sensitive to the data and it works well with symmterical data

- **Median** is better to use if data is **skewed (left, right)**

- **Median** is less affected by **outlier**

### Measure of Spread

##### quantify(express anything in nos) how much data values in a dataset vary or spread out from the center.

#### Types:
  * Range
  * IQR
  * Variance
  * Standard Deviation
  * Mean Absolute Deviation

  


**Measures of Spread** (Summary Statistics)

1.) It describes how spread apart or closed data points are

2.) There are some measure of spread (similarly like measure of center)

### **Variance** -> Average distance from each data point to the data's mean
  
   **Calculations of Variance**
   
   **Step 1** : Subtract mean from each data point
   
   ```
   distance = msleep["sleep_total] - np.mean(msleep["sleep_total])
   ```

   **Step 2 :** Square Each Distance
   ```
    sq_dists = distance ** 2
   ```

   **Step 3**:  Sum Squared Distance
   ```
    sum_sq_distances = np.sum(sq_dists)
   ```

   **Step 4:** Divide by no of data points - 1
   ```
    variance = sum_sq_distances / (83 -1)
   ```

#### Higher the variance the more spread data will be

or other way to calculate variance -

  ```
  np.var(msleep["sleep_total], ddof=1)
  ```

  -- without **ddof = 1** population variance is calculated instead of sample variance


### **Standard Deviation**
  
  * Another measure of spread
  * It is calcuated by taking square root of the variance

  ```
  np.sqrt(np.var(msleep["sleep_total], ddof=1))
  
  np.std(msleep["sleep_total], ddof=1)
  ```

### Mean Absolute Deviation

  * It takes absolute values of distance from Mean, then takes the mean of those distances

  ```
  distance  = msleep["sleep_total"] - mean(msleep["sleep_total"])
  np.mean(np.abs(distance))
  ```
#### Standard Deviation vs Mean Absolute Deviation

1.) Std dev squares distances, penalizing longer distances more than shorter ones

2.) Mean absolute deviation penalizes each distance equally

Quartiles:

Quartiles divide a dataset into four equal parts.
The first quartile (Q1) is the 25th percentile, meaning 25% of the data falls below this value.
The second quartile (Q2) is the median or the 50th percentile.
The third quartile (Q3) is the 75th percentile, meaning 75% of the data falls below this value.
Quintiles:

Quintiles divide a dataset into five equal parts.
Each quintile represents 20% of the data.
The first quintile (Q1) is the 20th percentile, the second quintile (Q2) is the 40th percentile, and so on.
Deciles:

Deciles divide a dataset into ten equal parts.
Each decile represents 10% of the data.
The first decile (D1) is the 10th percentile, the second decile (D2) is the 20th percentile, and so on.
These measures help in understanding the distribution and spread of the data.


## Interquartile Range (IQR)

3.) hieght of the box in a boxplot

### Outliers

1.) Data points that is substantially different from the others

2.) How do we know what a substantially difference is?

#### Rule

data_point < Q1 - 1.5 * IQR

data_point > Q3 + 1.5 * IQR

#### Process of finding out outliers

```
from scipy.stats import iqr
iqr = iqr(msleep["bodywt"])
lower_threshold = np.quantile(msleep["bodywt"], 0.25) - 1.5 * iqr
upper_threshold = np.quantile(msleep["bodywt"], 0.75) + 1.5 * iqr
```

```
# Subsetting for finding outlier
msleep[(msleep["bodywt"] < lower_threshold) | (msleep["bodywt"] < upper_threshold)]

```

### What is Distributions
