## Pandas
1.) Data Manipulation and Visualization library

### Pandas Method

1.) head() return first few rows of dataframe

2.) info() name of columns, no of values in the columns, no of missing values in the column, data type of the columns

3.) dataframe.shape -> row * columns

#### Grouped Summary Statistics

1.) dataframe.groupby("column_name1")["column_name2"].mean()

```
  Ex:
  dogs.groupby("color")["weight_kg"].mean()

  dogs.groupby("color")["weight_kg"].agg([min,max,sum])

  dogs.groupby(["color","breed"])["weight_kg"].mean()

  dogs.groupby(["color","breed"])["weight_kg","height_cm"].mean()
```

#### Pivot Tables

1.) It helps to calculate group summary statistics
similar to groupby

2.)dataframe.pivot_table(values="weight_kg", index="color")
  * values -> col to summarize
  * color -> col to **groupby**
  * **Bydefault pivot_table calculates mean() value of each group**

3.) dataframe.pivot_table(values="weight_kg", index="color", aggfunc=np.median)

4.) dataframe.pivot_table(values="weight_kg", index="color", aggfunc=[np.median,np.mean])

5.) groupby on multiple column

dataframe.pivot_table(values="weight_kg", index="color", columns="breed")
  * columns is used for additional col

6.) To overcome NaN value or missing value in pivot table

dataframe.pivot_table(values="weight_kg", index="color", columns="breed", fill_value=0)

7.) dataframe.pivot_table(values="weight_kg", index="color", columns="breed", fill_value=0, margins=True)

  * margins=True calculates each row and column mean at the end of rows and columns

````python
# Dataframe by rows
import pandas as pd

list_of_dictionaries = [
    {"name": "Arun", "age": 21, "gender": "M"},
    {"name": "Mohit", "age": 24, "gender": "M"}
    ]
new_dataframe = pd.DataFrame(list_of_dictionaries)

new_dataframe
````

````python
# Dataframe by column

dictionary_of_lists ={
    "name": ["Arun", "Mohit"],
    "age": [21, 24],
    "gender": ["M", "M"]
}

new_dataframe = pd.DataFrame(dictionary_of_lists)

new_dataframe.to_csv("students.csv")
````
